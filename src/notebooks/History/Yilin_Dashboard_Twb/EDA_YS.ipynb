```python
import pandas as pd
import numpy as np
```

```python
def tab_reader(filepathpattern,tabnum,skiprow,low_memory=False):
    tabs = []
    for i in range (tabnum):
        cur_index = i+1
        cur_filepath = filepathpattern.format(cur_index)
        try:
            tab = pd.read_csv(cur_filepath, skiprows=skiprow, low_memory=False)
        except:
            tab = pd.read_csv(cur_filepath, skiprows=skiprow,encoding='latin-1', low_memory=low_memory)
        tabs.append(tab)
    return tabs
```

```python
import altair as alt


m2017=alt.Chart(mb_r2017,title = "Current debt to equity in 2017").mark_bar().encode(

   alt.Y("North_American_Industry_Classification_System,_NAICS"
    ),
   alt.X("Current_debt_to_equity_(%)",
       ),
).properties(
   height=300, width=200
)

m2018=alt.Chart(mb_r2018,title = "Current debt to equity in 2018").mark_bar().encode(

   alt.Y("North_American_Industry_Classification_System,_NAICS"
    ),
   alt.X("Current_debt_to_equity_(%)",
       ),
).properties(
   height=300, width=200
)

m2019=alt.Chart(mb_r2019,title = "Current debt to equity in 2019").mark_bar().encode(

   alt.Y("North_American_Industry_Classification_System,_NAICS"
    ),
   alt.X("Current_debt_to_equity_(%)",
       ),
).properties(
   height=300, width=200
)
```

The current debt to equity is calculated as (current liabilitiesx100/equity), it is a measure of liquidity, which indicates a firm's relative ability to pay its short-term debts. The lower the positive ratio, the more liquid the business. We can see that for medium size business, the health care and social assistance, retail trade and construction rural business has a higher ratio, which means these type of business is not liquid. When we see the change from 2017 to 2019, we find that the Education services business in rural area has significantly lower its current debt to equity ratio, which means its improvement in liquidity. The information and cultural industries and management of companies are in good current debt to equity.

```python
alt.hconcat(m2017, m2018, m2019)
```

```python
mr2017=alt.Chart(mb_r2017,title = "Debt to equity ratio in 2017").mark_bar().encode(

   alt.Y("North_American_Industry_Classification_System,_NAICS"
    ),
   alt.X("Debt_to_equity_ratio_(times)",
       ),
).properties(
   height=300, width=200
)

mr2018=alt.Chart(mb_r2018,title = "Debt to equity ratio in 2018").mark_bar().encode(

   alt.Y("North_American_Industry_Classification_System,_NAICS"
    ),
   alt.X("Debt_to_equity_ratio_(times)",
       ),
   color = "Location_indicator",
).properties(
   height=300, width=200
)
mr2019=alt.Chart(mb_r2019,title = "Debt to equity ratio in 2019").mark_bar().encode(

   alt.Y("North_American_Industry_Classification_System,_NAICS"
    ),
   alt.X("Debt_to_equity_ratio_(times)",
       ),
   color = "Location_indicator",
).properties(
   height=300, width=200
)
```

The debt to equity ratio is calculated as total liabilities/total equity, it is a solvency ratio that indicates a firm's ability to pay its long-term debts. The lower the positive ratio, the more solvent the business. We can see that health care and social assistance of rural medium business still has a high debt to equity ratio. Information and cultural industry and management still perform well at a low debt to equity ratio. 

```python
alt.hconcat(mr2017, mr2018, mr2019)
```
