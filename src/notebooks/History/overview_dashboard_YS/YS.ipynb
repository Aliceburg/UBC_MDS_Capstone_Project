```python
def funct(prov_name):
    df_sub = df_selected[df_selected["GEO"]==prov_name]
    tit = "Rural population infomation of "+prov_name
    chart = (
        alt.Chart(df_sub,title=tit)
        .mark_bar()
        .encode(
            x=alt.X("content_name"),
            y="VALUE")
        .properties(height=250, width=800))
    return chart
```
