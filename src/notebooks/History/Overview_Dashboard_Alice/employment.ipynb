```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
employment_filepath = "employment.csv"
employ_data = pd.read_csv(employment_filepath)
```

```python
employ_break = employ_data.loc[employ_data['Lable'] == 'Total']
```

```python
employ_break
```

```python
employ_df = employ_data.melt(id_vars=['Lable'], var_name=['Year'], value_name='employment')
```

```python
employ_df
```

```python
employ_break_df = employ_break.melt(id_vars=['Lable'], var_name=['Year'], value_name='employment')
```

```python
employ_break_df
```

```python
a = alt.Chart(employ_df).mark_line(
    point=alt.OverlayMarkDef(color="blue")
).encode(
    x=alt.X('Year', title=None),
    y=alt.Y('employment', title='Employment (x 1000)'),
    color=alt.Color('Lable', legend=alt.Legend(
        orient='none', title=None,
        legendX=-10, legendY=-40,
        direction='horizontal',
        titleAnchor='middle',titleLimit=1000)),
    tooltip=['Lable', 'employment']
).properties(width=400, height=300).configure_legend(labelLimit= 0)

a
```

```python
b = alt.Chart(employ_break_df).mark_bar().encode(
    x=alt.X('Year', title=None),
    y=alt.Y('employment', title='Employment Population (x 1000)'),
    tooltip=['Lable','employment'],
    color=alt.Color('Lable', title=None)).properties(width=400, height=300)
b
```
