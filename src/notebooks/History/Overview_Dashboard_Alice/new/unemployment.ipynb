```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
def loaddata(filename, rateType):
    cur_filepath = filename
    data = pd.read_csv(cur_filepath)
    df = data.melt(var_name='Year',value_name='Unemployment rate(x1000)')
    df['Rate type'] = rateType
    return df
```

```python
unemploy_df = loaddata("unemployment.csv", "All population")
unemploy_df_metro = loaddata("unemployment_metro.csv", "Metropolitanagglomeration")
unemploy_df_nonmetro = loaddata("unemployment_Nonmetro.csv", "Non-metropolitan/non-agglomeration")
unemploy_df = pd.concat([unemploy_df,unemploy_df_metro,unemploy_df_nonmetro])
```

```python
alt.Chart(unemploy_df).mark_line(
    point=alt.OverlayMarkDef(color="blue")
).encode(
    x='Year',
    y='Unemployment rate(x1000)',
    color = alt.Color('Rate type', legend=alt.Legend(
        orient='none',
        legendX=-10, legendY=-40,
        direction='horizontal',
        titleAnchor='middle',titleLimit=1000)),
    tooltip=['Unemployment rate(x1000)','Rate type']
).properties(width=400, height=300).configure_legend(labelLimit= 0)
```
